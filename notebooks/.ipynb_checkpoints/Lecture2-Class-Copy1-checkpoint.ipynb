## 1.2 Why to Use NumPy?

* Python lists are excellent, general-purpose containers. They can be “heterogeneous”, meaning that they can contain elements of a variety of types, and they are quite fast when used to perform individual operations on a handful of elements.

* Depending on the characteristics of the data and the types of operations that need to be performed, other containers may be more appropriate; by exploiting these characteristics, we can improve speed, reduce memory consumption, and offer a high-level syntax for performing a variety of common processing tasks.

* NumPy shines when there are large quantities of “homogeneous” (same-type) data to be processed on the CPU.

## 1.11 Matrix Multiplication

If you start with these matrixes:

```python
a = np.array([[1, 2],
            [3, 4]])
b = np.array([[5, 6],
            [7, 8]])
```

For element-wise matrix multiplication, you can use multiply() function.

```python
arr_result = np.multiply(a, b)
arr_result
```

Let the matrices \( A \) and \( B \) be:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad
B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

The element-wise multiplication (Hadamard product) is defined as:

$$
A \circ B = \begin{bmatrix} 
1 \cdot 5 & 2 \cdot 6 \\ 
3 \cdot 7 & 4 \cdot 8 
\end{bmatrix}
= \begin{bmatrix} 5 & 12 \\ 21 & 32 \end{bmatrix}
$$

Thus, the result of `np.multiply(arr1, arr2)` is:

$$
\begin{bmatrix} 5 & 12 \\ 21 & 32 \end{bmatrix}
$$


For the matrix product of two arrays, you can use matmul() function.

```python
arr_result = np.matmul(a, b)
arr_result
```

Let the matrices \( A \) and \( B \) be:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad
B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

The matrix multiplication of \( A \) and \( B \) is defined as:

$$
A \times B =
\begin{bmatrix} 
(1 \cdot 5 + 2 \cdot 7) & (1 \cdot 6 + 2 \cdot 8) \\ 
(3 \cdot 5 + 4 \cdot 7) & (3 \cdot 6 + 4 \cdot 8) 
\end{bmatrix}
$$

Calculating each element:

$$
A \times B =
\begin{bmatrix} 
5 + 14 & 6 + 16 \\ 
15 + 28 & 18 + 32 
\end{bmatrix}
=
\begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}
$$

Thus, the result of `np.matmul(a, b)` is:

$$
\begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}
$$


For the dot product of two arrays, you can use dot() function.

```python
arr_result = np.dot(arr1, arr2)
arr_result
```

Let the matrices \( A \) and \( B \) be:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad
B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

The matrix multiplication of \( A \) and \( B \) is defined as:

Computing each element:

Thus, the result of `np.dot(arr1, arr2)` is:

## 3 Introduction to Data Preprocessing 


Data preprocessing is a crucial step in data science and machine learning, ensuring that raw data is cleaned, transformed, and prepared for analysis.

Real-world data is often incomplete, inconsistent, or contains errors, making preprocessing essential for improving model accuracy and efficiency.


## 3.1 Key Steps in Data Preprocessing

1. Data Collection – Gather raw data from various sources like databases, APIs, or files.
2. Data Cleaning – Handle missing values, remove duplicates, and correct errors.
3. Data Transformation – Normalize, scale, or encode categorical variables for better model performance.
4. Feature Engineering – Create new meaningful features or select the most relevant ones.
5. Data Splitting – Divide the dataset into training, validation, and test sets.

Proper data preprocessing enhances the quality of insights and ensures better decision-making in data-driven projects.

## 3.4 Feature Engineering

Now we will add new features to our dataset to gain more insights and also train our models

We can add moving averages and also time features

```python
df["MA7D_Close"] = df["Close/Last"].rolling(window=7, min_periods=1).mean()
df["MA30D_Volume"] = df["Volume"].rolling(window=30, min_periods=1).mean()

df["year"] = df.index.year
df["month"] = df.index.month
df["day"] = df.index.day
df["weekday"] = df.index.day_name()
df["weekday_numeric"] = df.index.weekday

df["is_weekend"] = df.index.weekday > 4
df.head(10)
```
